```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## plt.imshow

```python
img = cv2.imread('../data/cat.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

## line

```python
x, y, w, h = 180, 15, 150, 335
img_temp = img_rgb.copy()
cv2.line(img_temp, (x, y), (x+w, y+h), (200, 100, 200), 2)
plt.imshow(img_temp)
plt.show()
```

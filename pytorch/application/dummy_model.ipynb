```python
import torch, torchvision
```

```python
# %%script true
# Simple module for demonstration
class MyModule(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = torch.nn.Linear(4, 5)

    def forward(self, x):
        return self.linear(x).clamp(min=0.0, max=1.0)

model = MyModule()

dummy_input = torch.randn(1, 4)
torch.onnx.export(model, dummy_input, "test.onnx", output_names=['output'])
```

```python
# %%script true
model = torchvision.models.__dict__['inception_v3'](pretrained=True)
dummy_input = torch.randn(1, 3, 299, 299)
torch.onnx.export(model, dummy_input, "inception_v3.onnx", output_names=['output'])
```

## Inception

```python
from torchvision.models.inception import InceptionA

model = InceptionA(192, pool_features=32)
dummy_input = torch.randn(1, 192, 35, 35)
model(dummy_input)
torch.onnx.export(model, dummy_input, "InceptionA.onnx", output_names=['output'])
```

```python

from torchvision.models.inception import InceptionE

model = InceptionE(1280)
dummy_input = torch.randn(1, 1280, 8, 8)
model(dummy_input)
torch.onnx.export(model, dummy_input, "InceptionE.onnx", output_names=['output'])
```
